```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
digits=load_digits()
```

```python
print(digits)
```

```python
df=pd.DataFrame(digits.data)
df["Target"]=digits.target
print(df.head())
```

```python
kmeans=KMeans(n_clusters=10,random_state=42)
kmeans.fit(digits.data)
```

```python
df["Cluster"]=kmeans.labels_
print(df.head())
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(digits.data)
plt.scatter(pca_result[:,0],pca_result[:,1],c=df["Cluster"],cmap="Blues")
plt.xlabel("PCA Components 1")
plt.ylabel("PCA Components 2")
plt.title("K means clustering on Digits Dataset")
plt.show()
```

```python
cluster_centers=kmeans.cluster_centers_.reshape(10,8,8)
fig,axis=plt.subplots(2,5,figsize=(8,4))
for i,ax in enumerate(axis.flat):ax.imshow(cluster_centers[i],cmap="gray")
ax.set_title(f"Cluster {i}")
ax.axis("off")
plt.show()
```
